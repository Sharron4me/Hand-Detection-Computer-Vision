```python
import numpy as np
import cv2


def nothing(x):
    pass


cap = cv2.VideoCapture(0)
minH =10
maxH = 50
minS = 77
maxS = 212
minV = 115
maxV = 189

while 1:
    
    
    low = np.array([minH,minS,minV])
    high = np.array([maxH,maxS,maxV])
    
    ret, img = cap.read()
    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
    hsv = cv2.inRange(hsv,low,high)
    hsv = cv2.medianBlur(hsv,5)
    kernel=cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
    dilation = cv2.dilate(hsv,kernel,iterations = 1)
    image, contours, hierarchy = cv2.findContours(dilation,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    large_cnt= 0
    for i in range(len(contours)):
        if cv2.contourArea(contours[i]) > cv2.contourArea(contours[large_cnt]):
            large_cnt = i
    
    cv2.drawContours(img ,contours,large_cnt ,[0, 0, 255], 1 )
    cnt = contours[large_cnt]
    hull = cv2.convexHull(cnt,returnPoints = False)
    defects = cv2.convexityDefects(cnt,hull)

    for i in range(defects.shape[0]):
        s,e,f,d = defects[i,0]
        start = tuple(cnt[s][0])
        end = tuple(cnt[e][0])
        far = tuple(cnt[f][0])
        cv2.line(img,start,end,[0,255,0],2)
        cv2.circle(img,far,5,[0,0,255],-1)

    cv2.imshow('HSVIED',img)
    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break

cap.release()
cv2.destroyAllWindows()
```
